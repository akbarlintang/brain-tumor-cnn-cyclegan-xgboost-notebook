```python
# NEW YES DATA

def create_tfrecord(image_path, image_name, target):
    # Load the image
    with Image.open(image_path) as image:
        # Convert the image to bytes
        image_bytes = io.BytesIO()
        image = image.resize((256, 256))
        image.save(image_bytes, format='JPEG' if image.format == 'JPEG' else 'PNG')
        image_bytes = image_bytes.getvalue()
        # image_bytes = image.tobytes()

    # Create a TFRecord example
    example = example_pb2.Example()

    # Set the image name feature
    example.features.feature['image_name'].bytes_list.value.extend([image_name.encode()])

    # Set the image feature
    example.features.feature['image'].bytes_list.value.extend([image_bytes])

    # Set the target
    example.features.feature['target'].int64_list.value.append(target)

    # Serialize the example to a string
    serialized_example = example.SerializeToString()

    return serialized_example

# Example usage
dataset_path = '/content/drive/MyDrive/Data split skripsi/200/generated_data_1/yes/'  # Replace with your dataset folder path in Google Drive
tfrecord_path = '/content/drive/MyDrive/Data split skripsi/80-10-10/tfrecord/200_new_yes_data.tfrecord'  # Replace with the desired output TFRecord path in Google Drive

# List all image files in the dataset folder
file_list = [file for file in os.listdir(dataset_path) if file.lower().endswith(('.jpg', '.jpeg', '.png'))]

# Create the TFRecord writer
writer = tf.io.TFRecordWriter(tfrecord_path)

# Iterate over the image files and create TFRecord examples
for file_name in file_list:
    file_path = os.path.join(dataset_path, file_name)
    image_name = os.path.splitext(file_name)[0]
    target = 1  # Replace with the actual target/label value for the image

    # Create the TFRecord example
    serialized_example = create_tfrecord(file_path, image_name, target)

    # Write the example to the TFRecord file
    writer.write(serialized_example)

# Close the TFRecord writer
writer.close()
```
